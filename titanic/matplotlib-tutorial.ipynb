```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Simple labelled plot
t = np.arange(0.0, 4.0, 0.02)
s = 1 + np.cos(2 * np.pi* t)
fig, ax = plt.subplots()
ax.plot(t,s)
ax.set(xlabel='Xlabel', ylabel='Ylabel', title='Title')
ax.grid()
```

```python
# Multiple subplots
t1 = np.arange(0.0, 2.0, 0.02)
s1 = 1 + np.sin(2 * np.pi* t1)
t2 = np.arange(0.0, 2.0, 0.02)
s2 = 1 + np.cos(2 * np.pi* t2)

plt.subplot(1,2,1)
plt.plot(t1,s1, 'x')
plt.xlabel('X1 Label')
plt.ylabel('Y1 Label')

plt.subplot(1,2,2)
plt.plot(t2,s2, '+')
plt.xlabel('X2 Label')

plt.suptitle('A tale of 2 subplots')
```
